```python
from google_play_scraper import reviews, Sort
import pandas as pd
import time

APP_ID = "com.lazada.android"   # Lazada Android
TARGET = 30000                  # จำนวนรีวิวที่ต้องการดึง
BATCH_SIZE = 2000               # ขนาด batch ต่อรอบ (ไม่แนะนำให้เกิน 5000)

all_reviews = []
token = None

while len(all_reviews) < TARGET:
    print(f"กำลังดึงชุดที่ {len(all_reviews)} - {len(all_reviews) + BATCH_SIZE} ...")

    batch, token = reviews(
        APP_ID,
        lang='th',
        country='th',
        sort=Sort.NEWEST,
        count=BATCH_SIZE,
        continuation_token=token
    )

    if not batch:
        print("ไม่มีรีวิวเพิ่มแล้ว / ถูกจำกัดจาก Google")
        break

    all_reviews.extend(batch)

    # หน่วงเพื่อกันโดนบล็อก (สำคัญ!)
    time.sleep(1.2)

    # ถ้า token เป็น None แปลว่าไม่มีหน้าแล้ว
    if token is None:
        print("รีวิวหมดแล้ว ดึงต่อไม่ได้แล้ว")
        break

print("ดึงมาได้ทั้งหมด:", len(all_reviews))

# บันทึกไฟล์
df = pd.DataFrame(all_reviews)
df.to_csv("Lazada30.csv", index=False, encoding="utf-8-sig")
df.to_excel("Lazada30.xlsx", index=False)

print("บันทึกไฟล์เสร็จแล้ว!")
```
